# Bonus Excercise

Have you ever played **"Word Association Game"?**

It's quite simple!
I give you **word A** and **B**. Try to find the link between these two words, and **list the chain of words that starts from word A and leads up to word B**.

Let's say, I give you "`Sushi`"🍣 and "`desert`"🏜️.
You can say.... `Sushi -> fish -> sea -> beach -> sand -> desert`!!!

**How would you implement this game with word embeddings?**

First, we need to load word embeddings. You can train yours if you want. Or, you can use any existing pre-trained models.

```python
from gensim.models import word2vec
import gensim.downloader as api

info = api.info()  # show info about available models/datasets
model = api.load("glove-wiki-gigaword-100")  # download the model and return as object ready for use
```

Your task is to define a list of words that you're interested!
The list is going to be used to construct a graph, so it's recommended to keep it no more than 1,000 words for computational reasons.
As an idea, you can find the most common noun, adjective, etc.

```python
word_list = ['dog', 'cat', 'fish', 'sea', 'beach', 'sand', 'desert', 'mountain', 'river', 'forest', 'tree', 'bird', 'sky', 'cloud', 'rain', 'sun', 'moon', 'star', 'city', 'village', 'car', 'train', 'plane', 'food', 'pizza', 'sushi', 'japan', 'italy', 'cheese', 'bread', 'fruit', 'apple', 'banana', 'orange', 'grape', 'queen', 'king', 'prince', 'princess', 'throne', 'castle', 'mouse', 'rabbit', 'monkey', 'spider', 'puppy', 'kitten', 'pet', 'athletics', 'olympic', 'championships', 'swim', 'pool', 'golds', 'softball', 'volleyball', 'wife', 'mother', 'daughter', 'emperor', 'elizabeth', 'rome', 'milan', 'naples', 'venice', 'turin', 'vienna', 'genoa', 'berlin']
```

Here, we define a graph, and add nodes (words) and edges (similarities) using `networkx` library.
Since we will be using a shortest path algorithm that finds the path with the minimum score, we modify the similarity calculation a bit to compute "difference" rather than "similarity."

```python
import networkx as nx

# instantiate the graph
graph = nx.DiGraph()

# add nodes (words)
for x in word_list:
    graph.add_node(x)

# add edges (weights)
for x in word_list:
    for y in word_list:
        weight = pow(1 - model.similarity(x, y), 2) # modified similarity, or difference
        graph.add_edge(x, y, weight=weight)
```

Now, we define the function `association()` that produces the shortest path that goes from word A to word B.

```python
def association(node_1 , node_2):
    graph.remove_edge(node_1, node_2) # we remove the immediate edge that directly connects the two words
    data = nx.dijkstra_path(graph, node_1, node_2) # compute the shortest path using Dijkstra's algorithm
    weight = pow(1 - model.similarity(node_1, node_2), 2)
    graph.add_edge(node_1, node_2, weight=weight)
    for i, x in enumerate(data):
        print(f'{i}: {x}')
```

Let's try out!

```python
word_a = 'dog'
word_b = 'forest'

association(word_a, word_b)
```
