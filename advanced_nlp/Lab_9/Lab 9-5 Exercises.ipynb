## Retrieval Strategies

### 1) BM25 RAG (Baseline)

### 2) Hybrid RAG (BM25 + Dense + Reranker)

Implement a **three-stage retrieval pipeline**:

1. **BM25 retrieval**: Retrieve top-20 candidates using BM25
2. **Dense retrieval**: Retrieve top-20 candidates using a bi-encoder (e.g., `all-MiniLM-L6-v2`)
3. **Fusion**: Merge the two candidate sets using **Reciprocal Rank Fusion (RRF)**:

$$\text{RRF}(d) = \sum_{r \in \{\text{BM25}, \text{Dense}\}} \frac{1}{k + \text{rank}_r(d)}$$

   where $k = 60$.

4. **Rerank**: Take top-20 fused candidates and rerank with a cross-encoder (e.g., `cross-encoder/ms-marco-MiniLM-L-6-v2`)
5. **Select top-k** from the reranked list and feed to the LLM

## Evaluation

### Retrieval Metrics
1. **Recall@5**: Fraction of questions where the gold answer string appears in at least one of the top-5 retrieved passages
2. **MRR@5**: Mean Reciprocal Rank - average of $1/\text{rank}$ of the first passage containing the answer (0 if not found in top 5)

### Generation Metrics
3. **Token F1**: Average token-level F1 between predicted and gold answers
4. **Exact Match (EM)**: Fraction of predictions that exactly match the gold answer (after lowercasing and stripping whitespace)

### Efficiency
5. **Latency**: Average time per query (retrieval + generation) for each pipeline
